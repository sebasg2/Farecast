```python


def get_airport_coordinates(country):
    import airportsdata
    
    airports = airportsdata.load('IATA')
    iata_code = airports_dict.get(country)
    print(iata_code)
    if iata_code:
        airport = airports.get(iata_code)
        if airport:
            return airport['lat'], airport['lon']
    return None, None
```

```python
from math import radians, sin, cos, sqrt, atan2

def haversine(lat1, lon1, lat2, lon2):
    R = 6371  # Earth radius in kilometers
    dlat = radians(lat2 - lat1)
    dlon = radians(lon2 - lon1)
    a = sin(dlat/2)**2 + cos(radians(lat1)) * cos(radians(lat2)) * sin(dlon/2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))
    distance = R * c
    return distance
```

```python
def calculate_distance(row):
    
    origin_lat, origin_lon = get_airport_coordinates(row['Origin_Country'])
    destination_lat, destination_lon = get_airport_coordinates(row['Destination_Country'])
    if origin_lat is not None and destination_lat is not None:
        return haversine(origin_lat, origin_lon, destination_lat, destination_lon)
    return None
```

```python


def map_to_continent(country_name):
    
    from Countrydetails import countries
    
    country = countries.all_countries()
    continent_countries_dict=country.countries_in_continents()
    
    try:
        # Iterate through the dictionary to find the continent
        for continent, countries_list in continent_countries_dict.items():
            if country_name in countries_list:
                return continent
        return 'Unknown'  # Return 'Unknown' if country not found in the dictionary
    except Exception as e:
        print(f"Error mapping {country_name} to continent: {e}")
        return 'Unknown'
```

```python

def calculate_flight_duration(distance_km):
    # Velocidad promedio de un avión comercial en km/h
    average_speed_kmh = 900
    if distance_km is None:
        return None

    # Calcular la duración en horas
    duration_hours = distance_km / average_speed_kmh

    # Convertir la duración a decimal
    hours = int(duration_hours)
    minutes = (duration_hours - hours) * 60
    duration_decimal = hours + (minutes / 60)

    return round(duration_decimal, 2) 

```

```python
from datetime import datetime, timedelta
import numpy as np
import warnings

# Suppress all warnings
warnings.filterwarnings('ignore')

def calculate_distance_two(origin_country, destination_country):
    origin_lat, origin_lon = get_airport_coordinates(origin_country)
    destination_lat, destination_lon = get_airport_coordinates(destination_country)
    
    if origin_lat is not None and destination_lat is not None:
        return haversine(origin_lat, origin_lon, destination_lat, destination_lon)
    
    return None

def determine_if_stop_needed(distance):
    # Define the distance threshold for requiring a stop
    stop_threshold = 6437.376  # Example threshold in kilometers

    # Determine if a stop is needed based on distance
    if distance > stop_threshold:
        return True
    else:
        return False

def find_best_travel_time(origin, destination):
    best_month = None
    best_day = None
    best_hour = None

    # Initialize an empty list to store mean prices for each month
    mean_prices_by_month = []

    # Calculate date range from today to one year from now
    start_date = datetime.now()
    end_date = start_date + timedelta(days=365)  # One year from no6w

    # Representative days (1 per month)
    days_in_range = []
    for i in range(9):  # Adjust the number of representative days as needed
        # Calculate the day for each representative month
        month_date = start_date + timedelta(days=i * 30)  # Approximate days in a month
        days_in_range.append(month_date.day)

    hours_of_day = [8, 12, 18]  # Representative hours of departure

    # Iterate over months from current month to 12 months ahead
    for month_offset in range(0, 12):
        month = (start_date.month + month_offset - 1) % 12 + 1  # Ensure month wraps around correctly
        print(f"Checking month {month}...")

        month_mean_prices = []  # List to store mean prices for current month
        
        # Iterate over representative days
        for day in days_in_range:
            day_mean_prices = []  # List to store mean prices for current day
            
            # Iterate over hours of day
            for hour in hours_of_day:
                # Determine if stop is needed based on distance
                distance = calculate_distance_two(origin, destination)
                stops = 1 if determine_if_stop_needed(distance) else 0

                # Predict flight price for each combination
                predicted_price = predecir(Hour_of_Departure=hour, airline='', Stops=stops,
                                           Origin_Country=origin, Destination_Country=destination,
                                           day=day, month=month, year=start_date.year)

                # Calculate mean price for the specific day/hour combination
                mean_price = np.mean(predicted_price)
                day_mean_prices.append(mean_price)

                print(f"Checked day {day}, hour {hour} in month {month}... Mean Price: {mean_price}")

            # Store the mean price for the day if it's the best found so far in this month
            month_mean_prices.append(np.mean(day_mean_prices))
        
        # Store the mean prices for this month in the overall list
        mean_prices_by_month.append(np.mean(month_mean_prices))
        print("Mean price of the month ",month," ",np.mean(month_mean_prices))

        # Determine if this month has the lowest mean price
        if len(mean_prices_by_month) == 1 or mean_prices_by_month[-1] < np.min(mean_prices_by_month):
            best_month = month
            best_day_index = np.argmin(month_mean_prices)
            best_day = days_in_range[best_day_index]

            print(f"New best month found: {best_month}, Best Day: {best_day}, Mean Price: {mean_prices_by_month[-1]}")

    # Find the lowest mean price across all months
    lowest_mean_price = np.min(mean_prices_by_month)
    best_month_index = np.argmin(mean_prices_by_month)

    # Adjust best_month to ensure it is within 1-12 range
    best_month = (start_date.month + best_month_index) % 12 + 1
    best_year = start_date.year + (start_date.month + best_month_index - 1) // 12  # Adjust year if month wraps around

    # Find the best hour for the best day and best month
    day_mean_prices = []
    for hour in hours_of_day:
        # Predict flight price for each hour on the best day and month
        predicted_price = predecir(Hour_of_Departure=hour, airline='', Stops=stops,
                                   Origin_Country=origin, Destination_Country=destination,
                                   day=best_day, month=best_month, year=start_date.year)

        # Calculate mean price for the specific hour
        mean_price = np.mean(predicted_price)
        day_mean_prices.append(mean_price)

    # Determine the best hour
    best_hour_index = np.argmin(day_mean_prices)
    best_hour = hours_of_day[best_hour_index]
    lowest_price = day_mean_prices[best_hour_index]

    print(f"Best travel time found - Month: {best_month}, Year: {best_year}, Day: {best_day}, Hour: {best_hour}, Lowest Price: {lowest_price}")

    return best_month, best_day, best_hour, lowest_price

# Example usage
origin = 'Venezuela'
destination = 'China'

best_month, best_day, best_hour, lowest_price = find_best_travel_time(origin, destination)
print(f"Best month: {best_month}, Best day: {best_day}, Best hour: {best_hour}, Lowest price: {lowest_price}")




```
